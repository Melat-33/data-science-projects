```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Ensure data compatibility
X_train = np.array(X_train) if not isinstance(X_train, np.ndarray) else X_train
X_test = np.array(X_test) if not isinstance(X_test, np.ndarray) else X_test

# Instantiate classifiers
classifiers = [
    LogisticRegression(class_weight="balanced", random_state=1234),
    GaussianNB(),
    KNeighborsClassifier(),
    DecisionTreeClassifier(class_weight="balanced", random_state=1234),
    RandomForestClassifier(class_weight="balanced", random_state=1234),
]

# Define a result table as a DataFrame
result_table = pd.DataFrame(columns=["classifiers", "fpr", "tpr", "auc"])

# Train the models and predict the probabilities on the evaluation set
for cls in classifiers:
    model = cls.fit(X_train, y_train)
    yproba = model.predict_proba(X_test)[:, 1]  # Get probabilities for the positive class

    # Calculate ROC curve and AUC
    fpr, tpr, _ = roc_curve(y_test, yproba)
    auc = roc_auc_score(y_test, yproba)

    # Append results to the table
    result_table = pd.concat(
        [
            result_table,
            pd.DataFrame(
                {"classifiers": [cls.__class__.__name__], "fpr": [fpr], "tpr": [tpr], "auc": [auc]}
            ),
        ],
        ignore_index=True,
    )

# Set the name of the classifiers as index labels
result_table.set_index("classifiers", inplace=True)

# Display the result table
print(result_table)
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score, classification_report, confusion_matrix, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import pickle

def store_results(name, X_train, y_train, X_test, y_test, model, folds, norm="N/A", c="N/A", Class_weight="N/A"):
    '''
    Creates an entry to add to the results dataframe.
    
    Parameters:
    - name: Name of the model  
    - X_train: Training features
    - y_train: Training labels
    - X_test: Test features
    - y_test: Test labels
    - model: Model to be fit
    - folds: Number of folds in cross-validation
    - norm: L1 or L2 regularization
    - c: Inverse of regularization strength; must be a positive float.
    - Class_weight: Weights associated with classes.
    '''
    # Train the model
    model.fit(X_train, y_train)

    # Create the full filename with the .pkl extension
    filename = f"{name}.pkl"

    # Save the trained model
    with open(filename, 'wb') as f:
        pickle.dump(model, f)

    # Predict on the test dataset
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)

    # Compute metrics
    train_accuracy_score = accuracy_score(y_train, y_pred_train)
    test_accuracy_score = accuracy_score(y_test, y_pred)
    train_f1_score = round(f1_score(y_train, y_pred_train), 2)
    test_f1_score = round(f1_score(y_test, y_pred), 2)

    print('Classification Report:\n')
    print(classification_report(y_test, y_pred))
    print("=" * 100)

    # Compute and plot the confusion matrix
    cf_matrix = confusion_matrix(y_test, y_pred)
    categories = ['Negative', 'Positive']
    group_names = ['TN', 'FP', 'FN', 'TP']
    group_percentages = ['{0:.2%}'.format(value) for value in cf_matrix.flatten() / np.sum(cf_matrix)]
    labels = [f'{v1}: {v2}\n\n{v3}' for v1, v2, v3 in zip(group_names, cf_matrix.flatten(), group_percentages)]
    labels = np.asarray(labels).reshape(2, 2)
    sns.heatmap(cf_matrix, annot=labels, cmap='Blues', fmt='',
                xticklabels=categories, yticklabels=categories)
    plt.xlabel("Predicted Values", fontdict={'size': 14}, labelpad=10)
    plt.ylabel("Actual Values", fontdict={'size': 14}, labelpad=10)
    plt.title("Confusion Matrix", fontdict={'size': 18}, pad=20)
    plt.show()
    print("=" * 100)

    print('Train F1 Score:\n')
    print(round(f1_score(y_train, y_pred_train), 4))
    print("=" * 100)

    print('Test F1 Score:\n')
    print(round(f1_score(y_test, y_pred), 4))
    print("=" * 100)

    # Compute and plot the ROC curve
    y_proba = model.predict_proba(X_test)[:, 1]  # Probabilities for the positive class
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    auc_score = roc_auc_score(y_test, y_proba)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc_score:.2f})")
    plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
    plt.xlabel("False Positive Rate (FPR)")
    plt.ylabel("True Positive Rate (TPR)")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    plt.show()

    # Cross-validation scores
    auc = cross_val_score(model, X_train, y_train, cv=folds, n_jobs=-1, scoring='roc_auc').mean()
    accuracy = cross_val_score(model, X_train, y_train, cv=folds, n_jobs=-1).mean()
    scores = cross_val_score(model, X_train, y_train, cv=folds, n_jobs=-1, scoring='f1')
    f1score = scores.mean()

    # Create the result entry
    entry = {
        'Model': [name],
        'Class_weight': [Class_weight],
        'CrossVal_Mean(AUC)': [auc],
        'train_accuracy_score': [train_accuracy_score],
        'test_accuracy_score': [test_accuracy_score],
        'train_f1_score': [train_f1_score],
        'test_f1_score': [test_f1_score],
        'CrossVal_Mean(Accuracy)': [accuracy],
        'CrossVal_Mean(F1_score)': [f1score],
        'CrossVal1(f1)': [scores[0]],
        'CrossVal2(f1)': [scores[1]],
        'CrossVal3(f1)': [scores[2]],
        'CrossVal4(f1)': [scores[3]],
        'CrossVal5(f1)': [scores[4]],
        'Regularization': [norm],
        'C': [c]
    }

    result = pd.DataFrame(entry)
    return result
```

In the accuracy vs max_depth graph you can see that:

- After max_depth = 5, the training accuracy increased, whereas test accuracy decreased.

This means that if you continue increasing the depth:
- The accuracy for training data will reach 100%, but
- The accuracy for test data will reduce to a small value.

Therefore, in this case, the optimal value of max_depth parameter should be 5 based on Accuracy, let's take a look at f1_score

```python
#Define the parameters
parameters = {'max_depth': range(1,20)}

#You can define the number of folds for cross-validation as follows:
n_folds = 5

#Using the following code, call the decision tree classifier method:
dtree = DecisionTreeClassifier(criterion='gini', random_state=0)

#Use GridSearchCV to implement max_depth as follows:
tree=GridSearchCV(dtree,parameters,cv=n_folds,scoring="f1",return_train_score=True)

#Fit the training data in the Decision Tree as follows:
tree.fit(XNS_train, yNS_train)

#You can view the results as follows:
f1_scores = tree.cv_results_
```

```python
def print_recall_scores(ensemble, feats, true_labels):
    '''
    Prints the recall scores for base estimators in a sklearn ensemble model.
    '''
    scores = []
    for model_idx, model in enumerate(ensemble.estimators_):
        if model_idx == 0:
            print('='*40)
        preds = model.predict(feats)
        scores.append(recall_score(true_labels, preds))
        model_recall = np.round(recall_score(true_labels, preds), 5)
        print(f'Recall for Base Model {model_idx+1}:\t', model_recall)
        if model_idx < (len(ensemble.estimators_) - 1):
            print('-'*40)
        else:
            print('='*40)
    ensemble_preds = ensemble.predict(feats)
    print("Mean Recall Score:\t\t", np.round(np.array(scores).mean(), 5))
    print("Std Deviation:\t\t\t", np.round(np.array(scores).std(), 5))
    print("Range:\t\t\t\t", np.round(np.array(scores).ptp(), 5))
    print(f'Overall Recall for model:\t {np.round(recall_score(y_test, ensemble_preds), 5)}')
```
